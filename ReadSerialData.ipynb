## 金融序列化数据读取
---
>句子形式

```python
trans_f = open(r'./data/transf.txt', 'r', encoding='utf-8')
```

```python
import json
```

```python
class SerialSentce:
    def __init__(self, text, entity_list):
        self.text = text
        self.length = len(text)
        if len(entity_list) < 1:
            self.entity_list = []
        else:
            self.entity_list = entity_list

    def display(self):
        print('text--->', self.text)
        print('length--->', self.length)
        print('entitys--->', self.entity_list)
```

```python
def json_to_obj(d):
    return SerialSentce(d['text'], d['entity_list'])
```

```python
def get_sent_list(trans_f):
    sent_list=[]
    line=trans_f.readline()
    while line:
        serial=json.loads(line,object_hook=json_to_obj)
        sent_list.append(serial)
        line=trans_f.readline()
    trans_f.close()
    return sent_list
```

```python
sent_list=get_sent_list(trans_f)
```

```python
len(sent_list)
```

```python
s=sent_list[10]
s.display()
```

```python
company_dic={}
project_dic={}
contract_dic={}
for s in sent_list:
    for e in s.entity_list:
        if 'B-jiafang' in e:
            company_dic[e]=e
        if 'B-yifang' in e:
            company_dic[e]=e
        if 'xiangmu' in e:
            project_dic[e]=e
        if 'hetong' in e:
            contract_dic[e]=e
```

```python
print('公司--->', len(company_dic))
print('项目--->', len(project_dic))
print('合同--->', len(contract_dic))
```

```python
for key in project_dic:
    print(key)
```

```python
for key in contract_dic:
    print(key)
```
