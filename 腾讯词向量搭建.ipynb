```python
from gensim.models import  KeyedVectors
```

```python
# 时间较久，大概要30分钟左右
tc_wv_model=KeyedVectors.load_word2vec_format('../data/Tencent_AILab_ChineseEmbedding.txt',binary=False)
```

```python
import json
```

```python
from collections import OrderedDict
```

```python
word_index=OrderedDict()
```

```python
for counter,key in enumerate(tc_wv_model.vocab.keys()):
    word_index[key]=counter
```

```python
with open('../data/tc_word_index.json','w') as fp:
    json.dump(word_index,fp)
```

```python
from annoy import AnnoyIndex
```

```python
tc_index=AnnoyIndex(200)
```

```python
i=0
for key in tc_wv_model.vocab.keys():
    v=tc_wv_model[key]
    tc_index.add_item(i,v)
    i+=1
```

```python
#时间较久
tc_index.build(10)
```

```python
tc_index.save('../data/tc_index_build10.index')
```

```python
# 准备一个反向id==>word映射词表
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
testItem=1
for item in tc_index.get_nns_by_item(word_index[u'自然语言处理'], 10):
    testItem=item
    print(reverse_word_index[item])
```
